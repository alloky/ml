```python
import torch
import torch.nn as nn
from torch.autograd import Variable
import torch.nn.functional as F
from torch.autograd import Variable
import torchvision
from torchvision.datasets import cifar  
import torchvision.transforms as transforms
```

# Классы цифара

```python
classes = ('plane', 'car', 'bird', 'cat',
           'deer', 'dog', 'frog', 'horse', 'ship', 'truck')
```

# transforms - преобразовывает PIL в pytorch tensor 

```python
transform = transforms.Compose(
    [transforms.ToTensor()])
```

# загружаем данные ставим параметр download=true

```python
train_dataset = cifar.CIFAR10('./', transform=transform, download=True)
```

Размер тренировочного датасета.
Если поставить флаг train=Falsе - получим тестовый датасет

```python
len(train_dataset)
```

В pytorch cуществует сущность - DataLoader. Он приннимает как аргумент объект Dataset и использует его для итерирования и организации батчей 

```python

trainloader = torch.utils.data.DataLoader(train_dataset, batch_size=16,
                                          shuffle=True, num_workers=0)
```
